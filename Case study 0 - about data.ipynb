```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
sns.get_dataset_names()
```

```python
df_tips = sns.load_dataset("tips")
df_tips
```

```python
df_tips["total_bill"]
#df_tips.total_bill
```

```python
df_tips.describe()
```

```python
df_tips.describe
```

```python
df_tips.corr(numeric_only = True)

#creating correlation on numeric data
#using by default pearson correlation formula
#shows highly related that is tightly coupled
#diagonaly it's 1 as every object is tightly coupled with itself
```

```python
sns.heatmap(df_tips.corr(numeric_only = True))

#shows correlation in colors
```

```python
sns.heatmap(df_tips.corr(numeric_only = True), cmap = "coolwarm", annot = True)
```

```python
df_tips.sex = df_tips.sex.map({"Female":1,"Male":0})
df_tips
```

```python
df_tips['smoker'] = df_tips['smoker'].map({"No":0, "Yes":1})
df_tips['day'] = df_tips['day'].map({"Thur":1, "Fri":2, "Sat":3, "Sun":4})
df_tips.time = df_tips.time.map({"Lunch":0, "Dinner":1})

df_tips
```

```python
sns.heatmap(df_tips.corr(), cmap = "coolwarm", annot = True)
```

```python
sns.set(style = 'whitegrid')
sns.scatterplot(x = "sex",
               y = 'tip',
               data = df_tips)
```

```python
sns.scatterplot(x = "total_bill",
               y = 'tip',
               data = df_tips)
```

```python
sns.relplot(x = "total_bill",
            y = 'tip',
            data = df_tips,
            col = 'sex',
           hue = 'smoker')
```

```python
sns.relplot(x = "total_bill",
            y = 'tip',
            data = df_tips,
           hue = 'day')
```

```python
sns.relplot(x = "total_bill",
            y = 'tip',
            data = df_tips,
           hue = 'day', 
           row = 'sex')
```

```python
sns.relplot(x = "total_bill",
            y = 'tip',
            data = df_tips,
           hue = 'day',
           col = 'time')
```

```python
plt.figure(figsize = (8, 6))
plt.plot(df_tips["day"], df_tips["total_bill"], marker = "o", linestyle = "--", color = "blue")


plt.title("Line Plot: Total Bill Across Days")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (8, 6))
plt.bar(df_tips["day"].unique(), df_tips.groupby("day")["total_bill"].mean(), color = "green")


plt.title("Bar Plot: Average Total Bill by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Average Total Bill ($)")
plt.show()
```

```python
plt.figure(figsize = (8, 6))
plt.scatter(df_tips["total_bill"], df_tips["tip"], color = "orange")


plt.title("Scatter Plot: Total Bill vs Tip") 
plt.xlabel("Total Bill (rs)")
plt.ylabel("Tip (rs)")
plt.show()
```
